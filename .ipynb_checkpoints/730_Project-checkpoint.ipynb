```python
# Data Cleaning and extraction
import pandas as pd
df = pd.read_csv('Electric_Vehicle_Population_Data.csv')
df = df[df['State'] == 'WA']
df = df.drop(['Electric Utility', 'Model'], axis=1)
#df = df.drop('veh_loc', axis=1)
df = df.dropna()
```

```python
#Extracting the coordinates
df['veh_loc_x'] = ''
df['veh_loc_y'] = ''
```

```python
for i in range(df.shape[0]):
    s = df.iloc[i,13]
    df.iloc[i, 15] = float(s[7:-1].split(' ')[0])
    df.iloc[i, 16] = float(s[7:-1].split(' ')[1])
```

```python
df = pd.read_csv('processed.csv')
```

```python
df1 = df.groupby(by='County', as_index = False).count()
county = df1['County'].values.tolist()
num_evs = df1['City'].values.tolist()
```

```python
costs = np.random.normal(3600, 10, len(county))
```

```python
num_evs_dict = {}
for i in range(len(county)):
  num_evs_dict[county[i]] = num_evs[i]
```

```python
num_stat = []
for i in num_evs:
  num_stat.append(max(1, int(round((i/25), 0))))
```

```python
num_stations = {}
for i in range(len(county)):
  num_stations[county[i]] = num_stat[i]
#*3000000/sum(num_evs)
```

```python
cost_dict = {}
for i in range(len(county)):
  cost_dict[county[i]] = costs[i]
```

```python
#@title Example

model = ConcreteModel()

# assuming number of charging stations is 15 for this county
#model.count_stations = Var(1, domain=NonNegativeIntegers)
model.loc_x = Var(range(15), domain=Reals)
model.loc_y = Var(range(15), domain=Reals)
xcoords = df[df['County'] == 'Columbia'].iloc[:,15].tolist()
ycoords = df[df['County'] == 'Columbia'].iloc[:,16].tolist()

# dist = 0
# for i in range(len(xcoords)):
#   for j in range(15):
#     dist = dist + math.sqrt((xcoords[i] - model.loc_x[j])**2 + (ycoords[i] - model.loc_y[j])**2)

# Objective
model.Objective = Objective(expr = sum((xcoords[i] - model.loc_x[j])**2 + (ycoords[i] - model.loc_y[j])**2 for i in range(len(xcoords)) for j in range(15)), sense=minimize)

# Constraints
model.Constraint1 = ConstraintList()
# Constraint 1: X-coord limits
for i in range(15):
  model.Constraint1.add(expr=model.loc_x[i] <= max(xcoords))
for i in range(15):
  model.Constraint1.add(expr=model.loc_y[i] <= max(ycoords))
for i in range(15):
  model.Constraint1.add(expr=model.loc_x[i] >= min(xcoords))
for i in range(15):
  model.Constraint1.add(expr=model.loc_y[i] >= max(ycoords))

# Constraint 2: 
#model.Constraint2 = Constraint(expr=model.count_stations >= 1)

opt = SolverFactory('ipopt', executable='/content/ipopt')
results = opt.solve(model, tee = True)
print(model.loc_x[0]())
```

```python
#@title Single County
county_name = 'Jefferson'
num_charge = num_stations[county_name]



model = ConcreteModel()

#model.count_stations = Var(1, domain=NonNegativeIntegers)
model.loc_x = Var(range(num_charge), domain=Reals)
model.loc_y = Var(range(num_charge), domain=Reals)
xcoords = df[df['County'] == county_name].iloc[:,15].tolist()
ycoords = df[df['County'] == county_name].iloc[:,16].tolist()

dist = 0
for i in range(len(xcoords)):
  for j in range(num_charge):
    dist += ((xcoords[i] - model.loc_x[j])**2 + (ycoords[i] - model.loc_y[j])**2)**0.5

# Objective
model.Objective = Objective(expr = dist, sense=minimize)

# Constraints
model.Constraint1 = ConstraintList()
# Constraint 1: X-coord limits
for i in range(num_charge):
  model.Constraint1.add(expr=model.loc_x[i] <= max(xcoords))
for i in range(num_charge):
  model.Constraint1.add(expr=model.loc_y[i] <= max(ycoords))
for i in range(num_charge):
  model.Constraint1.add(expr=model.loc_x[i] >= min(xcoords))
for i in range(num_charge):
  model.Constraint1.add(expr=model.loc_y[i] >= min(ycoords))

# Constraint 2: 
model.Constraint2 = ConstraintList()
for i in range(num_charge-1):
  model.Constraint2.add(expr=model.loc_x[i] - model.loc_x[i+1] >= 0.01)
  model.Constraint2.add(expr=model.loc_y[i] - model.loc_y[i+1] >= 0.015)

opt = SolverFactory('ipopt', executable='/content/ipopt')
#opt.solve(model) 
results = opt.solve(model, tee = False) 
print('County:', county_name)
x_stat = []
y_stat = []

for j in range(num_charge):
  x_stat.append(model.loc_x[i]())
  y_stat.append(model.loc_y[i]())
for i in range(num_charge):
  print('Station ', i+1)
  print('X Coordinate: ',  model.loc_x[i](), '\t', 'Y Coordinate: ',  model.loc_y[i]())
```
